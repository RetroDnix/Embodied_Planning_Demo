```python
from induce.induce_actions import induce
query = "帮我洗一下衣服"
response = """
def solution():
    # Step 1: 检查是否需要分类衣物，例如根据颜色或材质
    if not check("clothes_sorted"):
        execute("sort_clothes_by_color_or_material")  # 分类衣物

    # Step 2: 检查是否有洗衣机
    if not find("washing_machine"):
        execute("locate_washing_machine")  # 定位洗衣机

    # Step 3: 打开洗衣机并加入衣物
    if not check("washing_machine_open"):
        execute("open_washing_machine")  # 打开洗衣机
    execute("add_clothes_to_washing_machine")  # 将衣物加入洗衣机

    # Step 4: 加入洗衣液或洗衣粉
    if not check("detergent_added"):
        execute("add_detergent_to_washing_machine")  # 加入洗衣液

    # Step 5: 选择洗衣模式
    if not check("wash_mode_selected"):
        execute("select_wash_mode")  # 选择适合的洗衣模式

    # Step 6: 启动洗衣机
    execute("start_washing_machine")  # 启动洗衣机

    # Step 7: 等待洗衣机完成工作
    while not check("washing_completed"):
        wait(1000)  # 每隔一段时间检查是否完成

    # Step 8: 取出衣物
    execute("remove_clothes_from_washing_machine")  # 从洗衣机中取出衣物

    # Step 9: 晾晒衣物或使用烘干机
    if not find("clothes_drying_place"):  # 检查是否有晾衣场地
        execute("locate_clothes_drying_place")  # 定位晾衣场地

    if check("has_drying_machine"):
        execute("use_drying_machine_to_dry_clothes")  # 使用烘干机
    else:
        execute("hang_clothes_to_air_dry")  # 手动晾晒衣物

solution()
"""
induce(query, response)
```

```python
from induce.induce_actions import induce
query = "帮我收拾一下用餐后的桌面"
response = """
def solution():
    # Stage 1: Locate the dining table
    if not find("dining_table"):
        execute("search_for_table")
        wait(500)  # Wait to ensure table is located

    # Stage 2: Clear leftover food from the table
    while find("leftover_food"):
        execute("clear_leftover_food_from_table")

    # Stage 3: Remove plates and utensils from the table
    while find("plate") or find("utensil"):
        execute("remove_plate_from_table")
        execute("remove_utensils_from_table")

    # Stage 4: Wipe the table surface
    if find("dirty_surface"):
        execute("wipe_table_surface")

    # Stage 5: Arrange chairs if they are not well positioned
    if not check("chairs_are_well_positioned"):
        execute("arrange_chairs")

    # Stage 6: Inspect table to ensure cleanliness
    if not check("table_is_clean"):
        execute("repeat_cleaning_procedure")


solution()
"""
induce(query, response)
```
